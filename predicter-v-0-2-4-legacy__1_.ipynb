```python
model_path = '/kaggle/input/tyolo_022/pytorch/v.0.2.4/1/best.pt'
model = YOLOv10(model_path)
```

```python
df = pd.read_csv("/kaggle/input/sliver/sliver-ultrasound-detection/ssubmission.csv")
# /kaggle/input/sliver/test/test/116601.jpg
img_dir = '/kaggle/input/sliver/sliver-ultrasound-detection/test/test'
for idx,row in df.iterrows():               # idx : index, row: iloc
#     print(idx,"->",row)
    if pd.isna(row['Label'])==False:        # If there are data continue kup
#         print("Continue")
        continue
    img_file = str(row['Image File'])
    img_path = f'{img_dir}/{img_file}.jpg'
    print(img_path)
    results  = model(source = img_path, conf=0.25)     # Predict image by image
#     print(results)
    for result in results:
        if result.boxes:
            boxes  = result.boxes.xyxy.tolist()    # Tensor งงๆ -> [[x1,y1,z1,a1],[x2,y2,z2,a2],...[xn,yn,zn,an]]
            clases = [int(cls) for cls in result.boxes.cls.tolist()]    # [float] -> [int]
            print(f"file : {img_file} box : {boxes} class : {clases}")
            df.at[idx,'Annotation'] = boxes
            df.at[idx,'Label']      = clases
        else:
            df.at[idx,'Annotation'] = float('nan')
            df.at[idx,'Label']      = float('nan')
#     if idx==5:
#         break
    print()
```
