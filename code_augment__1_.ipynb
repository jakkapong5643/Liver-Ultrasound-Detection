```python
import os
import shutil

# ฟังก์ชันเพื่อดึงชื่อไฟล์จากโฟลเดอร์โดยไม่รวมส่วนขยาย
def get_file_names(folder_path, extension):
    return {os.path.splitext(f)[0] for f in os.listdir(folder_path) if f.endswith(extension)}

# กำหนดเส้นทางไปยังโฟลเดอร์ที่มีไฟล์ txt และ jpg
txt_folder = r'C:\Users\jakka\Downloads\Data_Raw\val\val\annotations'
jpg_folder = r'C:\Users\jakka\Downloads\Data_Raw\val\val\images'

# ดึงชื่อไฟล์จากโฟลเดอร์ txt และ jpg
txt_files = get_file_names(txt_folder, '.txt')
jpg_files = get_file_names(jpg_folder, '.jpg')

# หาไฟล์รูปที่ไม่ตรงกันระหว่างสองโฟลเดอร์
jpg_not_in_txt = jpg_files - txt_files

# กำหนดโฟลเดอร์ปลายทางที่จะย้ายไฟล์ jpg ไป
destination_folder = r'C:\Users\jakka\Downloads\Data_Raw\Out_train'

# ย้ายไฟล์ jpg ที่ไม่ตรงกันไปยังโฟลเดอร์ปลายทาง
for file_name in jpg_not_in_txt:
    src_path = os.path.join(jpg_folder, file_name + '.jpg')
    dst_path = os.path.join(destination_folder, file_name + '.jpg')
    shutil.move(src_path, dst_path)
```

```python
import os
import shutil

# ระบุตำแหน่งของโฟเดอร์ที่มีไฟล์ jpg และ txt
jpg_folder = r"C:\Users\jakka\Downloads\Data_Raw\train\train\images"
txt_folder = r"C:\Users\jakka\Downloads\Data_Raw\train\train\labels"
destination_folder =r"C:\Users\jakka\Downloads\Data_Raw\val\val\images"

# สร้างโฟเดอร์ปลายทางหากยังไม่มี
if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)

# ดึงชื่อของไฟล์ jpg และ txt ในโฟเดอร์แต่ละรอบ
jpg_files = os.listdir(jpg_folder)
txt_files = os.listdir(txt_folder)

# วนลูปทุกไฟล์ txt และย้ายไฟล์ txt และ jpg ที่ชื่อไม่ตรงกันไปยังโฟเดอร์ปลายทาง
for txt_file in txt_files:
    if txt_file.endswith(".txt"):
        jpg_name = os.path.splitext(txt_file)[0] + ".jpg"
        if jpg_name not in jpg_files:
            shutil.move(os.path.join(txt_folder, txt_file), os.path.join(destination_folder, txt_file))
            jpg_file = os.path.join(jpg_folder, jpg_name)
            if os.path.exists(jpg_file):
                shutil.move(jpg_file, os.path.join(destination_folder, jpg_name))
```

# Augmentation Data Train

```python
import os
from imgaug import augmenters as iaa
import cv2

# กำหนดโฟลเดอร์ของภาพต้นฉบับและโฟลเดอร์ที่จะเก็บภาพที่ถูก augment
input_dir = r"C:\Users\jakka\Downloads\Data_Raw\train\train\images"
output_dir = r"C:\Users\jakka\Downloads\Data_Raw\Aug\images"

# สร้างโฟลเดอร์ที่จะเก็บภาพที่ถูก augment ถ้ายังไม่มี
os.makedirs(output_dir, exist_ok=True)

# กำหนดจำนวนรูปที่เพิ่มได้ (รูปที่ถูก augment จากแต่ละรูปต้นฉบับ)
num_augmentations_per_image = 5

# กำหนดค่าสำหรับการ augment ภาพ
seq = iaa.Sequential([
    iaa.GaussianBlur(sigma=(0, 3.0))  # เพิ่ม noise ละเบียดกับภาพด้วยการใช้ Gaussian blur
])

# วนลูปผ่านภาพในโฟลเดอร์ต้นฉบับและทำ augmentation
for filename in os.listdir(input_dir):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        image_path = os.path.join(input_dir, filename)
        image = cv2.imread(image_path)
        images_aug = [seq.augment_image(image) for _ in range(num_augmentations_per_image)]  # ทำ augmentation ภาพตามจำนวนที่กำหนด

        # บันทึกภาพที่ถูก augment ไว้ในโฟลเดอร์ output
        for i, img_aug in enumerate(images_aug):
            output_path = os.path.join(output_dir, f"{os.path.splitext(filename)[0]}_aug_{i}.jpg")
            cv2.imwrite(output_path, img_aug)
```

```python
import cv2
import os
from imgaug import augmenters as iaa

# กำหนด path ของโฟลเดอร์รูปภาพและโฟลเดอร์ข้อความ
images_folder = r"C:\Users\jakka\Downloads\Data_Raw\train\train\images"
labels_folder = r"C:\Users\jakka\Downloads\Data_Raw\train\train\labels"

# กำหนด path ของโฟลเดอร์สำหรับเก็บไฟล์ที่ผ่านการ augment
augmented_images_folder = r"C:\Users\jakka\Downloads\Data_Raw\Aug\images"
augmented_labels_folder = r"C:\Users\jakka\Downloads\Data_Raw\Aug\labels"

# สร้างโฟลเดอร์ถ้ายังไม่มี
os.makedirs(augmented_images_folder, exist_ok=True)
os.makedirs(augmented_labels_folder, exist_ok=True)

# กำหนดจำนวนการ augment ที่ต้องการ
num_augmentations = 3  # ตัวอย่างเช่น เพิ่ม 5 ภาพต่อ 1 ภาพต้นฉบับ

# สร้างตัว augmenter
seq = iaa.Sequential([
    iaa.Multiply((0.5, 0.8), per_channel=0.5),  # Brightness
    iaa.AddToHueAndSaturation((-25, 25)),  # Color Jittering
    iaa.Sometimes(0.5, iaa.AddToHueAndSaturation((-10, 10))),  # Color Swapping
    iaa.AdditiveGaussianNoise(scale=(0.0, 0.10 * 75)),  # Adding Gaussian Noise
    iaa.Sometimes(0.5, iaa.GaussianBlur(sigma=(0.0, 0.5))),
], random_order=True)

# ลูปผ่านไฟล์ในโฟลเดอร์รูปภาพ
for filename in os.listdir(images_folder):
    if filename.endswith(".jpg"):
        # กำหนด path ของรูปภาพและข้อความ
        image_path = os.path.join(images_folder, filename)
        txt_filename = os.path.splitext(filename)[0] + ".txt"
        txt_path = os.path.join(labels_folder, txt_filename)
        
        # ตรวจสอบว่ามีไฟล์ข้อความหรือไม่
        if not os.path.exists(txt_path):
            print(f"Label file for {filename} not found, skipping.")
            continue
        
        # โหลดรูปภาพ
        image = cv2.imread(image_path)

        # อ่านข้อความจากไฟล์ txt
        with open(txt_path, 'r') as file:
            text_data = file.read()

        # ทำการ augmentation ของรูปภาพ
        for i in range(num_augmentations):
            augmented_image = seq.augment_image(image)

            # บันทึกภาพที่ผ่านการ augmentation
            augmented_image_path = os.path.join(augmented_images_folder, f"augmented_{i}_{filename}")
            cv2.imwrite(augmented_image_path, augmented_image)

            # บันทึกข้อความ
            augmented_text_path = os.path.join(augmented_labels_folder, f"augmented_{i}_{txt_filename}")
            with open(augmented_text_path, 'w') as file:
                file.write(text_data)

print("Augmentation complete.")
```
